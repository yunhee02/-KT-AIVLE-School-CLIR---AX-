```python
import cv2
import os

def extract_frames(video_path, output_dir, target_fps=10):
    # 비디오 파일 열기
    cap = cv2.VideoCapture(video_path)

    if not cap.isOpened():
        print("❌ 비디오를 열 수 없습니다.")
        return

    # 원본 비디오 FPS 및 총 프레임 수
    video_fps = cap.get(cv2.CAP_PROP_FPS)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    duration = total_frames / video_fps

    print(f"🎥 원본 FPS: {video_fps:.2f}, 총 프레임: {total_frames}, 길이: {duration:.2f}초")

    # 프레임 간격 계산
    frame_interval = int(video_fps / target_fps)

    # 출력 디렉토리 생성
    os.makedirs(output_dir, exist_ok=True)

    count = 0
    saved_count = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        if count % frame_interval == 0:
            frame_name = os.path.join(output_dir, f"frame_{saved_count:05d}.jpg")
            cv2.imwrite(frame_name, frame)
            saved_count += 1

        count += 1

    cap.release()
    print(f"✅ 총 {saved_count}장의 프레임을 저장했습니다.")

# 예시 사용
video_file = "/content/[ZZ_2_F1_20_56_D_241207] 돌발상황영상 K-City 낙하물 20m 주간(48)_측면_투기.mp4"         # 변환할 영상 파일명
output_folder = "frames_output"          # 저장 폴더
extract_frames(video_file, output_folder, target_fps=10)
```

# 촬영 이미지 (판교, 에버랜드)


```python
import cv2
import os
import numpy as np

def extract_frames(video_path, output_dir, target_fps=10):
    cap = cv2.VideoCapture(video_path)

    if not cap.isOpened():
        print(f"❌ 비디오를 열 수 없습니다: {video_path}")
        return

    video_fps = cap.get(cv2.CAP_PROP_FPS)
    frame_interval = int(video_fps / target_fps)

    basename = os.path.splitext(os.path.basename(video_path))[0]
    count = 0
    saved_count = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        if count % frame_interval == 0:
            frame_name = os.path.join(output_dir, f"{basename}_frame_{saved_count:05d}.jpg")

            # ✅ 한글/특수문자 경로 호환 저장 방식
            is_success, buffer = cv2.imencode(".jpg", frame)
            if is_success:
                with open(frame_name, "wb") as f:
                    f.write(buffer)
                saved_count += 1

        count += 1

    cap.release()
    print(f"✅ {basename}: 총 {saved_count}장의 프레임을 저장했습니다.")

def process_all_videos_in_folder(folder_path, output_dir, target_fps=10):
    os.makedirs(output_dir, exist_ok=True)
    for filename in os.listdir(folder_path):
        if filename.lower().endswith(".mp4"):
            video_path = os.path.join(folder_path, filename)
            extract_frames(video_path, output_dir, target_fps)

# 사용 예시 (raw string으로 경로 안전 처리)
video_folder = r"C:\Users\User\Downloads\video_ours"
output_folder = r"C:\Users\User\Downloads\frames_output2"
process_all_videos_in_folder(video_folder, output_folder, target_fps=10)
```

```python
import shutil
import os

def zip_folder(folder_path, zip_path):
    if not os.path.exists(folder_path):
        print(f"❌ 폴더가 존재하지 않습니다: {folder_path}")
        return
    shutil.make_archive(zip_path, 'zip', folder_path)
    print(f"✅ 폴더가 압축되었습니다: {zip_path}.zip")

# 사용 예시
folder_to_zip = "frames_output"
zip_output_path = "frames_output_zip"  # 확장자는 자동으로 붙음
zip_folder(folder_to_zip, zip_output_path)
```

# 0.5초당 1 이미지 생성 코드 (ITS 공공 데이터)

```python
import cv2
import os

def extract_frames_every_half_second(video_path, output_dir):
    cap = cv2.VideoCapture(video_path)

    if not cap.isOpened():
        print(f"❌ 비디오를 열 수 없습니다: {video_path}")
        return

    video_fps = cap.get(cv2.CAP_PROP_FPS)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    duration = total_frames / video_fps

    frame_interval = int(video_fps * 0.5)  # 0.5초마다 저장

    basename = os.path.splitext(os.path.basename(video_path))[0]
    count = 0
    saved_count = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        if count % frame_interval == 0:
            frame_name = os.path.join(output_dir, f"{basename}_frame_{saved_count:05d}.jpg")

            # ✅ 한글 경로 대응 저장 방식
            success, buffer = cv2.imencode(".jpg", frame)
            if success:
                with open(frame_name, "wb") as f:
                    f.write(buffer)
                saved_count += 1

        count += 1

    cap.release()
    print(f"✅ {basename}: 총 {saved_count}장의 프레임을 저장했습니다.")

def process_all_videos_in_folder(folder_path, output_dir):
    os.makedirs(output_dir, exist_ok=True)
    for filename in os.listdir(folder_path):
        if filename.lower().endswith(".mp4"):
            video_path = os.path.join(folder_path, filename)
            extract_frames_every_half_second(video_path, output_dir)

# 사용 예시
video_folder = r"C:\Users\User\Downloads\video_ITS"              # 🎥 영상 폴더
output_folder = r"C:\Users\User\Downloads\frames_output_ITS"     # 🖼️ 프레임 저장 폴더
process_all_videos_in_folder(video_folder, output_folder)
```
